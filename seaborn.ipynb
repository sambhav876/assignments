```python
#question 2

import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Step 2: Create a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Step 3: Customize the plot
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event", loc="upper right", bbox_to_anchor=(1.15, 1))

# Step 4: Show the plot
plt.show()
```
